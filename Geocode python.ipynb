```python
keys = [
    #"80402923a7454df6a2ab43cd13171cfb" # lalo
    # "df0e883dd29a43b4ba7a1b39cf40f12a" # esteban
    #"84b20f7974a24b60b4acc1f0416d8795", # eduardo
    #"16f1e15a620d483f97b60a0a5baa0e1d", # marquez
    #"b53d1aa5d3d34373ac023d90e3e1b891" # fuentes
    # "8bb08353bb5a4c3b82eae9de3d44b33e" # mia
    #"77781c9bcca943b78903c352e699335d" #nose238
     "67a01badffdf43719188d8f74e062f11" #J
    # "1ad62d607a85481bb063bbdbac473600" # JS
]
```

```python
states = []
start = 6616
end = 2500 + 6616
for key in keys:
    geocoder = OpenCageGeocode(key)
    for i in range(start, end+1):
        if i > len(df):
            break
        result = geocoder.reverse_geocode(df.iloc(0)[i]["latitud"], df.iloc(0)[i]["longitud"])
        try:
            r = result[0]["components"]["state"]
        except:
            r = ""
        states.append(r)
        print("element added: " + str(i) )
    start = end + 1
    end = start + end
    if start > len(df):
        break
```
