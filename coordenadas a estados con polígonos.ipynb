# Función que devuelve el estado en el que se encuentran un par de coordenadas

```python
# Recibe de parámetro [latitud, longitud]
def dentro_de_estado(coords):
    for i in range(0, len(estados)):
        polygon = shape(Polygon(estados[i]))
        coords = Point(coords)
        if polygon.contains(coords):
            return estados_str[i]
    return "Desconocido"
```

# Función para recibir y devolver un DF. Usarla en tableau

```python
def aggEstado(df):
    df['estados'] = df.apply(lambda row : dentro_de_estado([row['headers.xlongitud'], row['headers.xlatitud']]), axis = 1)
    return df
```
